```python
import pandas as pd
df = pd.read_csv('salaries_by_college_major.csv')
```

```python
clean_df = df.dropna()
clean_df.tail()
```

```python
clean_df['Mid-Career Median Salary'].max()
```

```python

x=clean_df['Mid-Career Median Salary'].idxmax()
```

```python
clean_df['Undergraduate Major'].loc[x]
```

```python
highest = clean_df['Mid-Career Median Salary'].idxmax()
high_proff = clean_df['Undergraduate Major'].loc[highest]
print("Profession earning most is : ",high_proff)
lowest = clean_df['Mid-Career Median Salary'].idxmin()
lowest_proff  = clean_df['Undergraduate Major'].loc[lowest]
print("Profession earning least : ",lowest_proff)

lowest_mid = clean_df["Mid-Career Median Salary"].idxmin()
lowest_mid_career = clean_df["Undergraduate Major"].loc[lowest_mid]
print(lowest_mid_career)

```

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1,'Spread',spread_col)
clean_df.head()
```

```python
low_risk = clean_df.sort_values("Spread")
low_risk[["Undergraduate Major","Spread"]].head()
```

```python
highest_paid = clean_df.sort_values("Mid-Career 90th Percentile Salary",ascending=False)
highest_paid_jobs = highest_paid[["Undergraduate Major","Mid-Career 90th Percentile Salary"]].head()

greatest_spread = clean_df.sort_values("Spread",ascending=False)
max_gap = greatest_spread[["Undergraduate Major","Spread"]].head()

highest_paid_jobs

```
